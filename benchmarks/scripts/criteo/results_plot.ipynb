```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("nord-dark")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["figure.dpi"] = 100
```

```python
df = pd.read_csv("criteo_results_halfday_cloud10.csv").set_index("dpu ndpu")
```

```python
(df["intel total time"]/df["dpu build time"]).plot(xlim=0, ylim=0)
plt.title("DPU acceleration factor against Intelex")
plt.xlabel("DPU cores")
plt.ylabel("Speedup")
plt.show()
```

```python
(df["sklearn total time"]/df["dpu build time"]).plot(xlim=0, ylim=0)
plt.title("DPU acceleration factor against sklearn")
plt.xlabel("DPU cores")
plt.ylabel("Speedup")
plt.show()
```
